```python
#this code imports sqlite3 and prints a statement stating it has been imported
import sqlite3
print("sqlite3 imported")
```

```python
# This creates a brand-new, empty database file named 'AdCampaigns2026.db'
conn = sqlite3.connect('AdCampaigns2026.db')
print("database created")
```

```python
# This line is required so Colab stops acting like Python and acts more like a SQL database terminal
%sql sqlite:///AdCampaigns2026.db
```

```python
%%sql
CREATE TABLE Marketing_Brands (
brand_id INTEGER PRIMARY KEY,
brand_name TEXT,
brand_type TEXT,
brand_budget INTEGER
) STRICT;
```

```python
%%sql
INSERT INTO Marketing_Brands (brand_id, brand_name, brand_type, brand_budget)
VALUES
(1, 'Doritos', 'Snacks', 7000000),
(2, 'AWS', 'Technology', 15000000),
(3, 'Pepsi', 'Beverages', 2000000),
(4, 'He_Gets_Us', 'Religious', 1000000),
(5, 'Dr_Pepper', 'Beverages', 5000000),
(6, 'IBM', 'Technology', 100000),
(7, 'Google', 'Technology', 8000000);
```

```python
%%sql
SELECT * FROM Marketing_Brands;
```

```python
%%sql
SELECT brand_name, brand_type FROM Marketing_Brands WHERE brand_budget > 700000;
```

```python
%%sql
SELECT brand_type FROM Marketing_Brands WHERE brand_type = 'Religious';
```

```python
%%sql
SELECT brand_name FROM Marketing_Brands WHERE brand_budget < 1000000;
```

```python
%%sql
SELECT * FROM Marketing_Brands WHERE brand_budget > 7000000
```

```python
%%sql
SELECT brand_name FROM Marketing_Brands WHERE brand_type IN ('Snacks', 'Beverages');
```

```python
%%sql
SELECT brand_name, brand_budget FROM Marketing_Brands ORDER BY brand_budget DESC;
```

```python
#Block 10
# Using SQLite3 (and all its magic) to create a
# table showing all the columns in ascending order by budget.
%%sql
SELECT * FROM Marketing_Brands ORDER BY brand_budget ASC;
```
